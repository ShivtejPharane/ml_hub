```python
import numpy as np


def sigmoid(x):
    return 1 / (1 + np.exp(-x))


def tanh(x):
    return np.tanh(x)


def relu(x):
    return np.maximum(0, x)


np.random.seed(0)
s_hrs = np.random.uniform(1, 11, 100)
score = 5 * s_hrs + np.random.normal(0, 2, 100)


s_hrs = (s_hrs - np.mean(s_hrs)) / np.std(s_hrs)
score = (score - np.mean(score)) / np.std(score)


w = np.random.randn()
b = np.random.randn()
learning_rate = 0.01
epochs = 1000


def gradient_descent(w, b, learning_rate, activation_func, epochs):
    losses = []
    for epoch in range(epochs):
        predictions = activation_func(w * s_hrs + b)
        error = predictions - score

        dw = np.mean(error * s_hrs)
        db = np.mean(error)

        w = w - learning_rate * dw
        b = b - learning_rate * db

        loss = np.mean(error**2)  # Mean Squared Error
        losses.append(loss)

        if epoch % 200 == 0:
            print(f"Epoch {epoch}: Loss = {loss}")

    print(f"Final values -> w: {w:.4f}, b: {b:.4f}")
    return w, b, losses


w_sigmoid, b_sigmoid, losses_sigmoid = gradient_descent(
    w, b, learning_rate, sigmoid, epochs
)
w_relu, b_relu, losses_relu = gradient_descent(w, b, learning_rate, relu, epochs)
w_tanh, b_tanh, losses_tanh = gradient_descent(w, b, learning_rate, tanh, epochs)
```

```python
#for temprature
import numpy as np
import matplotlib.pyplot as plt

def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def relu(x):
    return np.maximum(0, x)

def tanh(x):
    return np.tanh(x)

np.random.seed(0)
altitude = np.random.uniform(500, 4000, 100)  # Altitude in meters
temperature = 30 - 0.0065 * altitude + np.random.normal(0, 1, 100)  # Temperature decreases with altitude

altitude = (altitude - np.mean(altitude)) / np.std(altitude)
temperature = (temperature - np.mean(temperature)) / np.std(temperature)

w = np.random.randn()
b = np.random.randn()
learning_rate = 0.01
epochs = 1000

def gradient_descent(w, b, learning_rate, activation_func, epochs):
    losses = []
    for epoch in range(epochs):
        predictions = w * altitude + b
        error = predictions - temperature

        dw = np.mean(error * altitude)
        db = np.mean(error)

        w = w - learning_rate * dw
        b = b - learning_rate * db

        loss = np.mean(error**2)
        losses.append(loss)

        if epoch % 200 == 0:
            print(f"Epoch {epoch}: Loss = {loss}")

    print(f"Final values -> w: {w:.4f}, b: {b:.4f}")
    return w, b, losses

print("Training with Sigmoid Activation:")
w_sigmoid, b_sigmoid, losses_sigmoid = gradient_descent(w, b, learning_rate, sigmoid, epochs)

print("\nTraining with ReLU Activation:")
w_relu, b_relu, losses_relu = gradient_descent(w, b, learning_rate, relu, epochs)

print("\nTraining with Tanh Activation:")
w_tanh, b_tanh, losses_tanh = gradient_descent(w, b, learning_rate, tanh, epochs)

plt.plot(losses_sigmoid, label="Sigmoid")
plt.plot(losses_relu, label="ReLU")
plt.plot(losses_tanh, label="Tanh")
plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.title("Loss Curves for Different Activation Functions")
plt.legend()
plt.show()
```
