```python
#code for STUDENT PERFORMANCE PREDICTION
import numpy as np
study_hrs=np.array([1,2,3,4,5])
marks=np.array([55,60,65,70,75])
m=b=0
iterations=100
lrate=0.001
n=len(marks)
for i in range(iterations):
  marks_predicted=m*study_hrs+b
  cost=(1/n)*sum((marks-marks_predicted)**2)

  dm=(-2/n)*sum((marks-marks_predicted)*study_hrs)
  db=(-2/n)*sum((marks-marks_predicted))

  m=m-lrate*dm
  b=b-lrate*db
  print(f"m:{m} b:{b}  cost:{cost}")
```
