```python
#Code for linear regrassion
import numpy as np
x=np.array([1,2,3,4,5])
y=np.array([50,55,65,70,75])
m=0
b=0
n=len(x)
iterations=101
lrate=0.001
for iterations in range(iterations):
    y_predicted=m*x+b
    error=y-y_predicted
    cost=(1/n)*sum((y-y_predicted)*2)
    dm=-(2/n)*sum(error*x)
    db=-(2/n)*sum(error)
    m=m-lrate*dm
    b=b-lrate*db
    print(f"iterations{iterations} m:{m} b:{b} cost:{cost}")

```
